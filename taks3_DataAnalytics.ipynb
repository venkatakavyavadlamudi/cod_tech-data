```python
import pandas as pd

# Load the Iris dataset from GitHub

df = pd.read_csv("irisdataset.csv")

# View column names
print(df.columns)
```

```python
['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']


```

```python

df.rename(columns={'variety': 'species'}, inplace=True)
[{'label': col, 'value': col} for col in df.columns if col not in ['Id', 'Species']]
```

```python
options=[{'label': sp, 'value': sp} for sp in df['Species'].unique()],
value=df['Species'].unique().tolist()
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)

app.layout = html.Div([
    html.H2("ðŸŒ¸ Iris Species Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select X-axis Feature:"),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in df.columns if col != 'Species'],
            value='sepal_length'
        ),

        html.Label("Select Y-axis Feature:", style={'marginTop': '10px'}),
        dcc.Dropdown(
            id='y-axis',
            options=[{'label': col, 'value': col} for col in df.columns if col != 'Species'],
            value='sepal_width'
        ),
    ], style={'width': '45%', 'display': 'inline-block', 'padding': '20px'}),

    html.Div([
        html.Label("Filter by Species:"),
        dcc.Checklist(
            id='species-filter',
            options=[{'label': sp, 'value': sp} for sp in df['Species'].unique()],
            value=df['Species'].unique().tolist(),
            inline=True
        )
    ], style={'width': '50%', 'display': 'inline-block', 'padding': '20px'}),

    dcc.Graph(id='scatter-plot'),

    html.Hr(),

    html.H4("Species Count"),
    dcc.Graph(
        figure=px.histogram(df, x="Species", color="Species", title="Species Distribution")
    )
])
```

```python
@app.callback(
    Output('scatter-plot', 'figure'),
    Input('x-axis', 'value'),
    Input('y-axis', 'value'),
    Input('species-filter', 'value')
)
def update_plot(x_axis, y_axis, selected_species):
    filtered_df = df[df['Species'].isin(selected_species)]
    fig = px.scatter(
        filtered_df, x=x_axis, y=y_axis, color='species',
        title=f'{y_axis} vs {x_axis}', template='plotly_dark', symbol='species'
    )
    return fig
```

```python
app.run(debug=True)
```
